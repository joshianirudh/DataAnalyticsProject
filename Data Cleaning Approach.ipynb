*From the above data we see that a lot of tuples have attributes with missing attribute values. So it's time to clean it*

```python
# Checking for the number of unique attribute values in each attribute
temp = list()
for i in df.columns:
    temp.append((df[i].nunique(), i))
    
temp.sort()
temp
#(num unique values, column name)
```

*We see that Country Code and Country Name have significantly less number of attributes values compared to other attributes*

```python
print(df['Country Name'].unique())
print(df['Country Code'].unique())

#Both of them have values of only one country or NaN values so we remove these attributes inplace

df.drop('Country Name', inplace = True, axis = 1)
df.drop('Country Code', inplace = True, axis = 1)
```

```python
df.isnull().sum(axis = 0) # Checks number of null values in each column
```

*As we can see, each attribute has only 5 null values, which correspond to the last 5 tuples that are empty as seen using df.tail()*

```python
df.drop(df.tail(5).index, inplace = True)
```

```python
df.isnull().sum(axis = 0) # Checking again, we see we have no null values. Or do we?
```

*Looking at the data again, we see that several attributes have attribute values such as ".." or "...". We must also clear them*

```python
temp_df = np.transpose(df)
temp_df
```

```python
percentages = list()
for i in temp_df.columns:
    count = 0
    for j in temp_df[i]:
        if j != '..':
            count += 1
    percentages.append((count*100/temp_df.shape[0], i))
    
percentages.sort()

for i in percentages:
    if i[0] < 50:
        temp_df.drop(i[1], inplace = True, axis = 1)

temp_df = np.transpose(temp_df)
temp_df
```
