```python
#Importing all the necessary modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('new_output.csv')
```

```python
#Displaying the first 10 rows . Just to confirm the dataset = "new_output.csv"
df.head(10)
```

# Scatter plots

```python
#This is a dictionary containing some important features (Manually choosen) for which plots will be drawn.
d = {40:"GDP (current US$)", 42:"GDP per capita (constant 2010 US$)", 0:"Age-dependency ratio", 3:"Agriculture, forestry, and fishing, value added (% of GDP)", 9:"Broad money (% of GDP)",15:"Claims on central government",16:"Consumer price index ",19:"Domestic credit to private sector (% of GDP)", 22: "Exports of goods and services (% of GDP)", 27:"External balance on goods and services (% of GDP)",99:"Manufacturing, value added (% of GDP)",107:"Military expenditure (% of GDP)",   176:"Services, value added (% of GDP)", 186:"Trade (% of GDP)", 188:"Urban population (% of total population)"}        

for keys in d.keys():
    x = [i for i in range(1960,2020)]
    y = [df.iloc[keys,6:66]]
    plt.scatter(x, y, label= "Datapoints", color= "purple",marker= "o", s=10)
    plt.xlabel('x - axis')
    plt.ylabel('y - axis')
    plt.title(d[keys])
    plt.legend()
    plt.show()
```

# Line plots

```python
d = {40:"GDP (current US$)", 42:"GDP per capita (constant 2010 US$)", 0:"Age-dependency ratio", 3:"Agriculture, forestry, and fishing, value added (% of GDP)", 9:"Broad money (% of GDP)",15:"Claims on central government",16:"Consumer price index ",19:"Domestic credit to private sector (% of GDP)", 22: "Exports of goods and services (% of GDP)", 27:"External balance on goods and services (% of GDP)",99:"Manufacturing, value added (% of GDP)",107:"Military expenditure (% of GDP)",   176:"Services, value added (% of GDP)", 186:"Trade (% of GDP)", 188:"Urban population (% of total population)"}        

for keys in d.keys():
    x = np.array([i for i in range(1960,2020)])
    y = np.array([df.iloc[keys,6:66]])
    plt.plot(x, np.transpose(y), color='black', linestyle='-', linewidth = 3,marker='o', markerfacecolor='yellow', markersize=4)
    plt.xlabel("X-axis data")
    plt.ylabel("Y-axis data")
    plt.title(d[keys])
    plt.show()
```

# Box plots

Box plot for GDP (current US$)

```python
#Outlier markers in red
#Mean represented it in green
#Median is denoted by the line

red_circle = dict(markerfacecolor='red', marker='o')
mean_shape = dict(markerfacecolor='green', marker='D', markeredgecolor='green')


plt.boxplot(x=df.iloc[40,6:66], vert=False, flierprops=red_circle, showmeans=True, meanprops=mean_shape, );
```

Box plot for Urban population 
( Vertical Box plot )

```python
plt.boxplot(x=df.iloc[188,6:66], vert=True, flierprops=red_circle, showmeans=True, meanprops=mean_shape, );
```

Box plot for GDP per capita (constant 2010 US$)

```python
plt.boxplot(x=df.iloc[42,6:66], vert=False, flierprops=red_circle, showmeans=True, meanprops=mean_shape, );
```

Box plot for Military expenditure 


Instead of representing the median as a line, we can represent it by using a notch instead.

```python
plt.boxplot(x=df.iloc[107,6:66], vert=False, flierprops=red_circle, showmeans=True, meanprops=mean_shape,notch=True );

```
