```python
temp = list()
for i in df.columns:
    temp.append((df[i].nunique(), i))
    
temp.sort()
temp
```

```python
df.drop('Country Code', axis = 1, inplace = True)
df.drop('Country Name', axis = 1, inplace = True)
df.drop('Series Code', axis = 1, inplace = True)
```

```python
df.drop(df.tail(5).index, inplace = True)
```

```python
df.set_index("Series Name", inplace = True)
df = np.transpose(df)
```

```python
temp_df = df
percentages = list()
for i in temp_df.columns:
    count = 0
    for j in temp_df[i]:
        if j != '..':
            count += 1
    percentages.append((count*100/temp_df.shape[0], i))
    
percentages.sort()

for i in percentages:
    if i[0] < 100:
        temp_df.drop(i[1], inplace = True, axis = 1)

temp_df
```
